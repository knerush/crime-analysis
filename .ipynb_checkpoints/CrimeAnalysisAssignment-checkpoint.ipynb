```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# sns.set_style('whitegrid')
import datetime
%matplotlib inline
```

The first step we are making is data eye inspection to understand datasets structure, also adding couple of convenience columns e.g 'Date', 'Time Reported'

Observations:
Date and time are present in both sets, we can use those to depict weekly/day-night patterns

Structure is different in crime categories (potentially we can aggregate by criminal/non-criminal categories), 

City zones are present for Seattle, districts for SanFran, so we can conclude what are the safest city parts, can also display most frequent crime types per district/zone.

```python
df_seattle = pd.read_csv('seattle_incidents_summer_2014.csv',low_memory=False)
df_seattle['Time Reported'] = pd.to_datetime(df_seattle['Date Reported'])
df_seattle['Date'] = df_seattle['Time Reported'].values.astype('<M8[D]')
df_seattle.head()
```

```python
df_seattle['Date Reported'].head()
```

```python
df_sf = pd.read_csv('sanfrancisco_incidents_summer_2014.csv', parse_dates=True, infer_datetime_format=True)
df_sf['Date'] = pd.to_datetime(df_sf['Date'])
df_sf.head()
```

Ultimate question - which city is safer

Plan:
* compare absolute crime numbers in the cities (count per day)
* compare crime per capita, grouped by 7 days rolling average to reduce volatility, so visualization is more clear

* top 10 crime types


Conclusions:
1. visualise count per day/week - result - comparable avg crime number, per capita SF is safer
2. 

directions for further investigation:
* break down by crime type and destrict/sector
* day vs night, and fluctuations depending on week days
* crime severity (unify types into criminal-non-criminal)

```python
df_sf.groupby('Date')['IncidntNum'].count().plot(label = 'SF')
df_seattle[df_seattle['Date'] < datetime.date(2014, 9, 1)].groupby('Date')['General Offense Number'].count().plot(label = 'Seattle')
plt.legend()
```

```python
df_seattle['Summarized Offense Description'].unique()
```

```python
df_sf['Category'].unique()
```

```python
df_sf[df_sf['Category'] != 'NON-CRIMINAL'].groupby('Date')['IncidntNum'].count().plot(label = 'SF')
df_seattle[df_seattle['Date'] < datetime.date(2014, 9, 1)].groupby('Date')['General Offense Number'].count().plot(label = 'Seattle')
plt.legend()
plt.title('Crimes per day')
```

Seattle population 608,660, SF - 870,887 (source - wiki articles) SanFran population is 1,43 times bigger than Seatle
Below is visualization for 7 day moving average, for Crimes per capita for both cities

```python
sf_grouped = (df_sf.groupby('Date')['IncidntNum'].count()/1.43)
sf_grouped.rolling(7).mean().plot(label = 'SF')
seatlle_grouped = df_seattle[df_seattle['Date'] < datetime.date(2014, 9, 1)].groupby('Date')['General Offense Number'].count()
seatlle_grouped.rolling(7).mean().plot(label = 'Seattle')
plt.legend()
plt.title('Crimes per capita, 7 days rolling average')
```

Looks like SF is significantly safer, lets try to judge on crime severeties
1. top 10 crime categories
2. top 10 safest districts

```python
top10_seattle = df_seattle.groupby('Summarized Offense Description')['RMS CDW ID'].count().sort_values().tail(20)
fig, ax = plt.subplots(figsize = (10, 8))
top10_seattle.plot(kind = 'barh', ax = ax)
plt.title('Seattle - top 10 most occuring crimes')
```

```python
top10_sf = df_sf.groupby('Category')['IncidntNum'].count().sort_values().tail(20)
fig, ax = plt.subplots(figsize = (10, 8))
top10_sf.plot(kind = 'barh', ax = ax)
plt.title('San Francisco - top 10 most occuring crimes')
```
